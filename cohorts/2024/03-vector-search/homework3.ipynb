```python
class VectorSearchEngine():
    def __init__(self, documents, embeddings):
        self.documents = documents
        self.embeddings = embeddings

    def search(self, v_query, num_results=10):
        scores = self.embeddings.dot(v_query)
        idx = np.argsort(-scores)[:num_results]
        return [self.documents[i] for i in idx]

search_engine = VectorSearchEngine(documents=ml_zoomcamp_data, embeddings=X)
search_engine.search(v, num_results=5)
```

```python
from elasticsearch import Elasticsearch

es_client = Elasticsearch('http://localhost:9200') 

index_settings = {
    "settings": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    },
    "mappings": {
        "properties": {
            "text": {"type": "text"},
            "section": {"type": "text"},
            "question": {"type": "text"},
            "course": {"type": "keyword"},
            "id": {"type": "keyword"},
            "question_text_vector": {
                "type": "dense_vector",
                "dims": 768,
                "index": True,
                "similarity": "cosine"
            },
        }
    }
}

index_name = "course-questions"

es_client.indices.delete(index=index_name, ignore_unavailable=True)
es_client.indices.create(index=index_name, body=index_settings)
```

```python
for doc in tqdm(ml_zoomcamp_data):
    question = doc['question']
    text = doc['text']
    qa_text = f'{question} {text}'

    doc['question_text_vector'] = embedding_model.encode(qa_text)

    es_client.index(index=index_name, document=doc)
```

```python
search_query = {
    "field" : "question_text_vector",
    "query_vector" : v,
    "k" : 5,
    "num_candidates" : 10000

}

es_results = es_client.search(
    index=index_name,
    knn=search_query,
    source=["text","section","question","course","id"]
)
```

```python
def elastic_search_knn(field, vector):
    search_query = {
    "field" : field,
    "query_vector" : vector,
    "k" : 5,
    "num_candidates" : 10000

    }   

    es_results = es_client.search(
        index=index_name,
        knn=search_query,
        source=["text","section","question","course","id"]
    )

    result_doc = []

    for hit in es_results['hits']['hits']:
        result_doc.append(hit['_source'])

    return result_doc
```

```python
def question_vector_knn(q):
    question = q['question']
    course = q['course']

    v_q = embedding_model.encode(question)

    return elastic_search_knn('question_text_vector', v_q)
```

```python
def evaluate(ground_truth, search_function):
    relevance_total = []

    for q in tqdm(ground_truth):
        doc_id = q['document']
        results = search_function(q)
        relevance = [d['id'] == doc_id for d in results]
        relevance_total.append(relevance)

    return {
        'hit_rate': hit_rate(relevance_total),
    }
```

```python
evaluate(ground_truth, question_vector_knn)
```
